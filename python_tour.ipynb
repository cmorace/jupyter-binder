## Classes

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def add(self, other_point):
        self.x += other_point.x
        self.y += other_point.y
        

a = Point(1, 2)
b = Point(3, 4)
a.add(b)
print("a = (", a.x, ",", a.y, ")")
```

## Conditional Statements

```python
age = input("enter your age:")
if age.isnumeric():
    age = int(age)
    if age < 2:
        print("you are a baby")
    elif age < 17:
        print("you are a child")
    else:
        print("you are an adult") 
else:
    print("expected a number but input was", age)
```

## Iteration

### While Loop

```python
i = 0
while i < 10:
    print(i, end=",")
    i += 1
print(i)
```

### For Loops

```python
print("range with 1 argument:")
for i in range(10):
    print(i, end=",")
    
print("\n\nrange with 2 arguments")
for i in range(5,10):
    print(i, end=",")
    
print("\n\nrange with 3 arguments:")
for i in range(1,10,2):
    print(i, end=",")
print("")
for i in range(10,1,-2):
    print(i, end=",")
```

### Matplotlib
Matplotlib is a package for visualization

```python
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize=(16,9), dpi=300)
ax = fig.add_subplot()
x = np.linspace(0, 2*np.pi, 30)
y = np.sin(x)
b = ax.bar(x, y, width=0.03)
```
